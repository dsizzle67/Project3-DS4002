```python
import pandas as pd

# The MetObjects dataset is too large for GitHub so we will only use MetObjects_FINAL, but this is the code to create MetObjects_FINAL
df = pd.read_csv('MetObjects.csv')
df
```

```python
# Drop all columns except 'Object Number', 'Title', 'Period', and 'Link Resource'
drop_columns = []
for col in df:
    if col != 'Object Number' and col != 'Title' and col != 'Period' and col != 'Link Resource':
        drop_columns.append(col)
df_small = df.drop(drop_columns, axis='columns')
df_small
```

```python
# Print Unique values in Period column
df_small = df_small[df_small['Period'].notna()]
print(len(df_small))
df_small['Period'].unique()
df_small['Period'].value_counts()
```

```python
# Display value counts as DataFrame with better formatting
period_counts = df_small['Period'].value_counts().to_frame()
period_counts.columns = ['Count']
print(f"Total unique periods: {len(period_counts)}")
print("\nTop 10 most common periods:")
print(period_counts.head(10))
```

```python
df_small.to_csv('MetObjects_FINAL.csv', index=False)
```
